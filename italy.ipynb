```python
import pandas as pd
from sqlalchemy import create_engine, inspect
csv_file = "Resources\italian_vaccination.csv"
italy_df = pd.read_csv(csv_file)
italy_df.head()
```

```python
italy_grouped = italy_df.groupby(['administration_date']).sum()

italy_grouped.head()
```

```python
italy_df=italy_grouped.drop(['first_dose', 'second_dose', 'previous infection', 'ISTAT_regional_code'], axis = 1)
italy_df.head()
```

```python
sum_column = italy_df["males"] + italy_df["females"]

italy_df["people"] = sum_column
italy_df.head()
```
